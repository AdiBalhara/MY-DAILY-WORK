```python
##Checking for null values
train.isnull()
```

```python
##Removing useless coloumns
train = train.drop(columns=['Unnamed: 0','first','last','street','trans_num','cc_num'])
```

```python
train['trans_date_trans_time'] = pd.to_datetime(train['trans_date_trans_time'])

train['hour'] = train['trans_date_trans_time'].dt.hour
train['day'] = train['trans_date_trans_time'].dt.day
train['month'] = train['trans_date_trans_time'].dt.month
```

```python
train['dob'] = pd.to_datetime(train['dob'])

train['age'] = 2026 - train['dob'].dt.year
```

```python
train.head()
```

```python
train = train.drop(columns=['trans_date_trans_time','dob'])
```

```python
train.isnull().sum()
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['gender'] = le.fit_transform(train['gender'])
train['category'] = le.fit_transform(train['category'])
train['job'] = le.fit_transform(train['job'])
train['merchant'] = le.fit_transform(train['merchant'])
```

```python
train.head()
```

```python
##Handle outliers

import numpy as np

train['amt'] = np.log1p(train['amt'])
```

```python
##Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_cols = ['amt','lat','long','city_pop','merch_lat','merch_long','age']

train[numerical_cols] = scaler.fit_transform(train[numerical_cols])
```

```python
##Label Encoding

from sklearn.preprocessing import LabelEncoder

categorical_cols = ['city','state']

le = LabelEncoder()

for col in categorical_cols:
    train[col] = le.fit_transform(train[col])
```

```python
##Handle Imbalanced data

from imblearn.over_sampling import SMOTE

X = train.drop('is_fraud', axis=1)
y = train['is_fraud']

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)
```
