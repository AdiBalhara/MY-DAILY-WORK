```python
import pandas as pd
```

```python
test = pd.read_csv("../data/Test.csv")
```

```python
print(test.shape)
```

```python
test.head()
```

```python
##Checking for null values
test.isnull()
```

```python
##Removing useless coloumns
test = test.drop(columns=['Unnamed: 0','first','last','street','trans_num','cc_num'])
```

```python
test['trans_date_trans_time'] = pd.to_datetime(test['trans_date_trans_time'])

test['hour'] = test['trans_date_trans_time'].dt.hour
test['day'] = test['trans_date_trans_time'].dt.day
test['month'] = test['trans_date_trans_time'].dt.month
```

```python
test['dob'] = pd.to_datetime(test['dob'])

test['age'] = 2026 - test['dob'].dt.year
```

```python
test.head()
```

```python
test = test.drop(columns=['trans_date_trans_time','dob'])
```

```python
test.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

test['gender'] = le.fit_transform(test['gender'])
test['category'] = le.fit_transform(test['category'])
test['job'] = le.fit_transform(test['job'])
test['merchant'] = le.fit_transform(test['merchant'])
```

```python
##Handle outliers

import numpy as np

test['amt'] = np.log1p(test['amt'])
```

```python
##Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_cols = ['amt','lat','long','city_pop','merch_lat','merch_long','age']

test[numerical_cols] = scaler.fit_transform(test[numerical_cols])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test['is_fraud'].value_counts()
```

```python
##Label Encoding

from sklearn.preprocessing import LabelEncoder

categorical_cols = ['city','state']

le = LabelEncoder()

for col in categorical_cols:
    test[col] = le.fit_transform(test[col])
```

```python
##Handle Imbalanced data

from imblearn.over_sampling import SMOTE

X = test.drop('is_fraud', axis=1)
y = test['is_fraud']

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)
```
